```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def readFromExcelToDataFrame():
    return pd.read_excel('synthetic_data_with_missing_values.xlsx', 'Sheet1', index_col=0, parse_dates=True)
```

```python
df = readFromExcelToDataFrame()
df.info
```

```python
#dataframe with interpolated (method: linear) values 
df_interpolate = df.interpolate(method ='linear', limit_direction ='forward', limit = 1)
```

```python
#1st plot data with missing values 
ts = df.set_index('Dates')
ts.plot(grid=True)
```

```python
#2nd plot with interpolated data 
ts = df_interpolate.set_index('Dates')
ts.plot(grid=True)
```
