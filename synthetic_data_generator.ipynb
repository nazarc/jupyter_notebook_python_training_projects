```python
import pandas as pd
import numpy as np
```

```python
def generate_dates(start_date, end_date, n, unit):
    diff_days = (end_date - start_date).days + 1 
    return pd.date_range(start = start_date, end = end_date )
```

```python
def shuffle_rows(df):
    return  df.sample(frac=1)
```

```python
np.random.seed(0)
start_date = pd.to_datetime('2021-01-01')
end_date = pd.to_datetime('2021-12-31')
n = 10 
```

```python
dates = generate_dates(start_date=start_date, 
                              end_date=end_date,
                              n=n,
                              unit='D' )
df = pd.DataFrame(dates)
print(df)
df2 = shuffle_rows(df=df)

print(df2)
```
