```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def readFromExcelToDataFrame():
    return pd.read_excel('data_samples/synthetic_data_with_missing_values.xlsx', 'Sheet1', index_col=0, parse_dates=True)
```

```python
df = readFromExcelToDataFrame()
df.info
```

```python
#dataframe with interpolated (method: linear) values 
df_interpolate_linear = df.interpolate(method ='linear', limit_direction ='forward', limit = 1)
#dataframe with interpolated (method: linear) values 
df_interpolate_quadratic = df.interpolate(method ='quadratic', limit_direction ='forward', limit = 1)
#dataframe with interpolated (method: linear) values 
df_interpolate_nearest = df.interpolate(method ='nearest', limit_direction ='forward', limit = 1)
```

```python
#1st plot data with missing values 
ts = df.set_index('Dates')
ts.plot(grid=True, title='Original data with missing values')
#2nd plot with interpolated data
ts = df_interpolate_linear.set_index('Dates')
ts.plot(grid=True, title='Interpolation method linear')
#3rd plot with interpolated data
ts = df_interpolate_quadratic.set_index('Dates')
ts.plot(grid=True, title='Interpolation method quadratic')
#4th plot with interpolated data
ts = df_interpolate_nearest.set_index('Dates')
ts.plot(grid=True, title='Interpolation method nearest')
```
